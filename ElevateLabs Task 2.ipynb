```python
############################# SECTION -1 (IMPORTING DATA SETS) #####################################################

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

############################# SECTION -2 (LOADING THE CLEANED DATASET) ##############################################
df=pd.read_csv("C:\\Users\\Nitin Dutt Majji\\Downloads\\B.TECH\\Elevate Labs\\cleaned_titanic.csv")

############################# SECTION -3 (SUMMARY STATISTICS & INFO) ################################################
print(df.describe())
print(df.info())

############################# SECTION -4 (DISTRIBUTION OF AGE) ######################################################
plt.figure(figsize=(10,5))
sns.histplot(df['Age'].dropna(),bins=30,kde=True)
plt.title("Distribution of Age")
plt.show()

############################# SECTION -5 (FARE DISTRIBUTION BY PASSENGER CLASS) #####################################
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare distribution by passenger class")
plt.show()

############################# SECTION -6 (CORRELATION MATRIX HEATMAP) ###############################################
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

############################# SECTION -7 (PAIRPLOT OF SELECTED FEATURES) ############################################
sns.pairplot(df[['Age','Fare','Pclass','Survived']].dropna(),hue='Survived')
plt.suptitle("Pairplot",y=1.02)
plt.show()

############################# SECTION -8 (SURVIVAL COUNT BY GENDER) #################################################
sns.countplot(x='Sex', hue='Survived',data=df)
plt.title("Survival Count by Gender")
plt.show()
```
